### Qus.-1: use IF ELSE and ELIF to write a program in Python for you report card.

```python
eng_mark = 70
hin_mark = 80
mat_mark = 98
che_mark = 55
phy_mark = 78
sport_mark = 80
outof_mark = 600
total_marks = eng_mark + hin_mark + mat_mark + che_mark + phy_mark + sport_mark
per = (total_marks/outof_mark)*100
if per >= 90:
    print("Grade: A+")
elif per >= 80:
    print("Grade: A")
elif per >= 70:
    print("Grade: B+")
elif per >= 60:
    print("Grade: B")
elif per >= 50:
    print("Grade: C+")
elif per >= 40:
    print("Grade: C")
else:
    print("Fail")
print(f"Total marks: {total_marks}")
print("Percent: %.2f " %per )
```

### Qus.-2: Use For Loop to Print Prime Numbers in between 1 to 1000

```python
lower = 1
upper = 1000

print("Prime numbers between", lower, "and", upper, "are:")

for num in range(lower, upper + 1):
   # all prime numbers are greater than 1
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num)

```

### Qus.-3:  Write a program for printing the tables from 1,10 using Nested For Loop

```python
for i in range(1,11):
        for j in range(1,11):
            mul = i*j
            print(mul, end=" ")
        print("\n")
```
